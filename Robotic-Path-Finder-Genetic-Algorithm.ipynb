```python
import random

def generate_obstacles(grid_width, grid_height, start, goal, max_obstacles):
    grid = [(i, j) for i in range(grid_height) for j in range(grid_width)]
    path = [start]

    current = start

    while current != goal:
        next_step = random.choice([(0, 1), (1, 0), (0, -1), (-1, 0)])
        new_position = (current[0] + next_step[0], current[1] + next_step[1])

        if (0 <= new_position[0] < grid_height and
            0 <= new_position[1] < grid_width and
            new_position not in path):
            path.append(new_position)
            current = new_position
        else:
            break

    free_cells = set(grid) - set(path)

    obstacles = random.sample(free_cells, min(max_obstacles, len(free_cells)))

    return obstacles,path

GRID_WIDTH = 4
GRID_HEIGHT = 4
start = (0, 0)
goal = (3, 3)
obstacles,path = generate_obstacles(GRID_WIDTH, GRID_HEIGHT, start, goal, max_obstacles=5)

print("Obstacles:", obstacles)
print(path)
```

```python
import random

GRID_WIDTH = 4
GRID_HEIGHT = 4

def generate_obstacles(grid_width, grid_height, start, goal, max_obstacles):
    grid = [(i, j) for i in range(grid_height) for j in range(grid_width)]

    path = [start]
    current = start

    while current != goal:
        next_step = random.choice([(0, 1), (1, 0), (0, -1), (-1, 0)])
        new_position = (current[0] + next_step[0], current[1] + next_step[1])

        if (0 <= new_position[0] < grid_height and
            0 <= new_position[1] < grid_width and
            new_position not in path):
            path.append(new_position)
            current = new_position

    free_cells = set(grid) - set(path)
    obstacles = random.sample(free_cells, min(max_obstacles, len(free_cells)))

    return obstacles

OBSTACLES = generate_obstacles(GRID_WIDTH, GRID_HEIGHT, (0, 0), (3, 3), max_obstacles=5)

def fitness(path, goal):
    if any(step in OBSTACLES for step in path):
        return float('inf')

    distance_to_goal = abs(path[-1][0] - goal[0]) + abs(path[-1][1] - goal[1])
    return len(path) + distance_to_goal

def generate_random_path(start, goal):
    path = [start]
    while path[-1] != goal:
        current = path[-1]

        if current[0] < goal[0]:
            next_step = (1, 0)
        elif current[0] > goal[0]:
            next_step = (-1, 0)
        elif current[1] < goal[1]:
            next_step = (0, 1)
        else:
            next_step = (0, -1)

        new_position = (current[0] + next_step[0], current[1] + next_step[1])

        if (0 <= new_position[0] < GRID_HEIGHT and
            0 <= new_position[1] < GRID_WIDTH and
            new_position not in OBSTACLES and
            new_position not in path):
            path.append(new_position)

        if len(path) > GRID_WIDTH * GRID_HEIGHT:
            break
    return path

def crossover(parent1, parent2):
    crossover_point = random.randint(1, min(len(parent1), len(parent2)) - 1)
    return parent1[:crossover_point] + parent2[crossover_point:]

def mutate(path):
    mutation_point = random.randint(1, len(path) - 2)
    mutation_step = random.choice([(0, 1), (1, 0), (0, -1), (-1, 0)])
    new_position = (path[mutation_point][0] + mutation_step[0], path[mutation_point][1] + mutation_step[1])

    if (0 <= new_position[0] < GRID_HEIGHT and
        0 <= new_position[1] < GRID_WIDTH and
        new_position not in OBSTACLES):
        path[mutation_point] = new_position
    return path

def genetic_algorithm(start, goal, population_size, generations):
    population = []
    for _ in range(population_size):
        path = generate_random_path(start, goal)
        print(f"Generated path: {path}")
        population.append(path)

    for generation in range(generations):
        population = sorted(population, key=lambda path: fitness(path, goal))

        print(f"Generation {generation}: Best path = {population[0]}")

        if fitness(population[0], goal) == len(population[0]):
            return population[0]

        next_generation = population[:population_size // 2]

        while len(next_generation) < population_size:
            parent1, parent2 = random.choices(population[:10], k=2)
            child = crossover(parent1, parent2)
            child = mutate(child)
            next_generation.append(child)

        population = next_generation

    return population[0]


start = (0, 0)
goal = (3, 3)

best_path = genetic_algorithm(start, goal, population_size=50, generations=100)
print("Obstacles:", OBSTACLES)
print("Best Path found:", best_path)
```
